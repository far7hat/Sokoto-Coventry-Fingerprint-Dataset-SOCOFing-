```python
import os
from PIL import Image

def get_image_sizes(directory):
    sizes = []
    for subdir, _, files in os.walk(directory):
        for file in files:
            if file.lower().endswith('bmp'):
                file_path = os.path.join(subdir, file)
                with Image.open(file_path) as img:
                    sizes.append(img.size)
    return sizes

train_dir = train_easy_dir
image_sizes = get_image_sizes(train_dir)
print(image_sizes)

from collections import Counter

size_counter = Counter(image_sizes)
print(f"Most common sizes: {size_counter.most_common(5)}")

min_size = min(image_sizes, key=lambda x: x[0] * x[1])
max_size = max(image_sizes, key=lambda x: x[0] * x[1])

print(f"Smallest image size: {min_size}")
print(f"Largest image size: {max_size}")
```

```python
valid_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)

valid_train_gen = valid_datagen.flow_from_directory(
    validation_data_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical'
)
```

```python

class myCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        if logs.get('accuracy') is not None and logs.get('accuracy') > 0.98:
            print("\nReached 85% accuracy so cancelling training!") 
            self.model.stop_training = True
```

```python
def create_model():
    model = tf.keras.models.Sequential([
        # First convolution
        tf.keras.Input(shape=(150, 150, 3)),
        tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
        tf.keras.layers.MaxPooling2D(2, 2),
        # Second convolution
        tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
        tf.keras.layers.MaxPooling2D(2,2),
        # Third convolution
        tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
        tf.keras.layers.MaxPooling2D(2,2),
        # Fourth convolution
        tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
        tf.keras.layers.MaxPooling2D(2,2),
        # Flatten the results to feed into a DNN
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dropout(0.5),
        # 512 neuron hidden layer
        tf.keras.layers.Dense(512, activation='relu'),
        # Output layer
        tf.keras.layers.Dense(20, activation='softmax')  # Assuming 20 classes for age, gender, and ethnicity
    ])

    # Compile the model
    model.compile(optimizer='rmsprop',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model

# Create the model
model = create_model()
callbacks = myCallback()
```

```python
history = model.fit(
    easy_train_gen,
    epochs=40,
    validation_data=valid_train_gen,
    verbose=1,
    validation_steps=3,
    callbacks=[callbacks]
)

model.save_weights('easy_weights.h5')
```

```python
model = create_model()
model.load_weights('easy_weights.h5')

# Freeze the initial layers
for layer in model.layers[:-2]:  # Freezing all layers except the last two
    layer.trainable = False

# Compile the model again after freezing layers
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Continue training on the medium dataset
model.fit(medium_train_gen,    epochs=40,
    validation_data=valid_train_gen,
    verbose=1,
    validation_steps=3,
    callbacks=[callbacks])

# Save the model weights
model.save_weights('medium_weights.h5')
```
